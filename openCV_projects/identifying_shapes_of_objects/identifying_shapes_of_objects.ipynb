```python
import numpy as np
import cv2


#Getting image to observe
image = cv2.imread("shapes.png", cv2.IMREAD_COLOR)

cv2.imshow('Identifying shapes of objects', image)
cv2.waitKey(0)

#Preprocessing image to fit cv2.findContours
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, threshhold = cv2.threshold(gray_image, 125,255,1)

#We will return only external countorus..
#Chain approx none is set because we dont need all boundary points
contours, hiearchy = cv2.findContours(threshhold.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)


for cnt in contours:
    
    #Getting approximate polinoms to see of what shape object is
    #With lower second parameter, approximation will be more accurate
    approximate = cv2.approxPolyDP(cnt, 0.01*cv2.arcLength(cnt,True),True)

    if len(approximate) == 3:
            shape_name = "Triangle"
            cv2.drawContours(image, [cnt] ,0, (22,223,33), -1)
    #if approx lenght is 4, then it could be square or rectangle
    elif len(approximate) == 4:
            x,y,w,h = cv2.boundingRect(cnt)
            if abs(w-h) <=20:
                shape_name = "Square"
                cv2.drawContours(image, [cnt],0, (200,42,81), -1)
            else:
                shape_name = "Rectangle"
                cv2.drawContours(image,[cnt],0, (200,20,50), -1)
    
    elif len(approximate) == 10:
            shape_name = "Star"
            cv2.drawContours(image,[cnt], 0 , (20,77,244), -1)
    
    elif len(approximate) == 5:
            shape_name = "Pentagon"
            cv2.drawContours(image,[cnt], 0 , (44,120,244), -1)
    
    else: 
            shape_name = "Circle"
            cv2.drawContours(image, [cnt], 0 , (255,230,230), -1)
            
    #now we need to center the text holder
    M = cv2.moments(cnt)
    cx = int(M['m10'] / M['m00'])
    cy = int(M['m01'] / M['m00'])
    cv2.putText(image, shape_name, (cx-40, cy), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,0), 2)

    cv2.imshow('Identifying shapes of objects', image)
    cv2.waitKey(0)

cv2.destroyAllWindows()
```
